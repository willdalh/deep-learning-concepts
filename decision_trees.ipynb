```python
from sklearn import tree
X = [[0, 0], [0, 1], [1, 0], [1, 1], [1, 1]]
Y = [0, 1, 1, 0, 0]
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y)
```

```python
clf.predict([[0., 0.]])
```

```python
tree.plot_tree(clf, rounded=True, label="all", node_ids=False, impurity=False, feature_names=["x1", "x2"])
```
